```python
from sklearn.feature_extraction.text import TfidfVectorizer

import sys
sys.path.insert(0, '..')
from utils.read_data import read_articles_file
```

```python
article_texts_path = "/work3/s204163/wiki/article_texts"
article_ids_path = "/work3/s204163/wiki/article_ids"

N = 10
articles = read_articles_file(article_texts_path, N = N)
```

```python
corpus = [0] * N

for i, (id, (title, text)) in enumerate(articles.items()):
    corpus[i] = text
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(corpus)

out = vectorizer.get_feature_names_out().tolist()
out.reverse()
print(out)
```
